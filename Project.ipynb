# Introduction

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
dataset = pd.read_csv('tmdb_movies_data.csv')
```

# Data Wrangling

Total Number of Rows

```python
dataset
```

finding total number of NAN values in the Dataset in each column

```python
dataset.isna().sum()
```

# Data Cleaning

 Dropping Rows that have NaN values

```python
dataset.dropna(inplace=True)
```

Making New Index

```python
dataset.reset_index(inplace=True)
```

Dropping the two id and index column because we have new column as an index

```python
dataset.drop(['id','index'],axis=1)
```

```python
dataset.drop_duplicates(inplace=True)
```

# Exploratory Data Analysis 

```python
dataset.info()
```

Question: Which genres are most popular from year to year?

```python
dataset['release_date'] = pd.to_datetime(dataset.release_date)
```

```python
dataset['release_date'].dt.year
```

```python
final = dataset.sort_values('popularity',ascending=False)
```

```python
final = dataset.head(30)
```

```python
plt.figure(figsize=(20,20))
sns.scatterplot(final['release_date'].dt.year,final['popularity'],hue=final['genres'])
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol=1)
plt.show()
```

We can see that genre "Family|Animation|Adventure|Comedy" is the most popular genre among others in the 2015. We can see that in the 19th century, No genre became most popular as compared to "Family|Animation|Adventure|Comedy" genre in the 21 century.

Question : What kinds of properties are associated with movies that have high revenues?

#Sorting the Data in descending order of Revenue(from highest to lowest)

```python
final_2 = dataset.sort_values('revenue',ascending=False)
```

Selecting the top 40 Movies with Highest Revenues

```python
final_2 = final_2.head(40)
```

```python
final_2
```

Visualizing which Movie generated the highest Revenue

```python
plt.figure(figsize=(20,20))
sns.barplot(final_2['original_title'],final_2['revenue'])
plt.xticks(rotation='90',fontsize=15)
plt.yticks(fontsize=15)
plt.ticklabel_format(style='plain', axis='y')
plt.xlabel('Movie Title',fontsize=20)
plt.ylabel('Revenue Generated',fontsize=20)
plt.tight_layout()
plt.show()
```

The Above Visual shows the movies that generated highest revenue as compared to others. We can see that movie "AVATAR" generated the highest amount of revenue as compared to other movies. The above visual shows the top 40 movies that generated high revenue.

Analyzing the Relation between Vote Count and High revenue Generated Movies

```python
plt.figure(figsize=(10,10))
sns.scatterplot(final_2['original_title'],final_2['vote_count'],hue=final_2['revenue'])
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol=1)
plt.xticks(rotation='90',fontsize=10)
plt.yticks(fontsize=10)
plt.xlabel('Movie Title',fontsize=15)
plt.ylabel('Vote Counts',fontsize=15)
plt.tight_layout()
plt.show()
```

From the above figure, we can conclude that movies with highest revenues have large amount of vote counts given by the people. People actually voted more for movies that generated high revenue. So we can say that increase in votes increases the revenue of the movie and vice versa.
